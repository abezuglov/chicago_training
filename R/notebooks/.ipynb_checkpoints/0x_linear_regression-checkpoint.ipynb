Is there a relationship between advertising budget and the sales?

```r
#par(mfrow = c(1,3), pty='s')
scatter.smooth(x = adv$TV, y = adv$Sales, main = 'TV ~ Sales')
scatter.smooth(x = adv$Radio, y = adv$Sales, main = 'Radio ~ Sales')
scatter.smooth(x = adv$Newspaper, y = adv$Sales, main = 'Newspaper ~ Sales')
```

Yes, it looks like there is dependency...

How can we model the dependency? Perhaps we should start with a single variable, i.e. univariate model: Sales ~ TV

$\hat{y} = \beta_0 + \beta_1*TV$, where $\hat{y}$ is a model of Sales

Try to find the best values of $\beta_0$ and $\beta_1$ below:

```r
beta_0 <- 7.03
beta_1 <- 0.04
y_hat <- beta_0 + adv$TV*beta_1
plot(y = adv$Sales, x = adv$TV)
lines(y = y_hat, x = adv$TV, col = 'red')
```

Let us see visualize the errors:

```r
y_hat <- beta_0 + adv$TV*beta_1
err <- adv$Sales - y_hat
plot(density(err), main = 'Error distribution')
qqnorm(err)
#shapiro.test(err)
```

Is there a better way to estimate the coefficients rather than guessing?

### The Least Squares Approach

The least squares approach provides a closed form solution. The idea here is to minimize the cost function:

$J = \frac{1}{n}\sum_i \left[ y_i-\hat{y}_i\right]^2$, where $\hat{y}_i = \beta_0 + \beta_1*x_i$

According to the least squares approach, the equations for $\hat{\beta}_1$ and $\hat{\beta}_0$ are: 

$\hat{\beta}_1 = \frac{\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^{n}(x_i-\bar{x})^2}$

$\hat{\beta}_0 = \bar{y}-\hat{\beta}_1\bar{x}$

```r
beta_1_hat <- sum((adv$TV-mean(adv$TV))*(adv$Sales-mean(adv$Sales)))/sum((adv$TV-mean(adv$TV))**2)
beta_0_hat <- mean(adv$Sales-beta_1_hat*mean(adv$TV))
beta_0_hat
beta_1_hat
```

### Gradient Descent Approach

Another approach for finding linear regression parameters is based on finding error gradients $\frac{\partial{J}}{\beta_0}$ and $\frac{\partial{J}}{\beta_1}$ and updating the coefficients:

```r
beta_0 <- 0
beta_1 <- 0
num_iterations <- 10000
lambda1 <- 0.005
lambda2 <- 0.00005

for (i in 1:num_iterations){
    y_hat <- beta_0 + beta_1*adv$TV
    dy_hat <- y_hat-adv$Sales
    dLb0 <- mean(dy_hat) # error gradients
    dLb1 <- mean(dy_hat*adv$TV)

    beta_0 =  beta_0 - lambda1 * dLb0
    beta_1 =  beta_1 - lambda2 * dLb1
}

print(beta_0)
print(beta_1)
```

### Linear regression with R tools

Finally, using R tools:

```r
lr <- lm(Sales ~ TV, data = adv)
lr
```

## Assessing the Accuracy of the Coefficients

The assumption is that the *true* relationship between TV and Sales takes the form $Sales = f(TV)+\epsilon$, where $\epsilon$ is a mean-zero random error term:

$Sales = \beta_0+\beta_1*TV+\epsilon$

Run the code below to see what $\beta_0$ and $\beta_1$ are like for subsets of the original data:

```r
beta_0 <- c()
beta_1 <- c()
num_samples <- 10000
num_rows <- nrow(adv)
num_rows_sample <- num_rows*0.1

for (i in 1:num_subsets){
    adv_sample <- adv[sample(num_rows, num_rows_sample, replace = FALSE),]
    lr <- lm(Sales ~ TV, data = adv_sample)
    beta_0[i] <- coef(lr)[1]
    beta_1[i] <- coef(lr)[2]
}
#beta_0
#beta_1
```

Both coefficients may take extreme values:

```r
min(beta_0)
max(beta_0)
min(beta_1)
max(beta_1)
```

What do the possible models look like? 

```r
plot(y = adv$Sales, x = adv$TV)
for (i in 1:num_samples){
    y_hat = beta_0[i] + beta_1[i]*adv$TV
    lines(y = y_hat, x = adv$TV, col = 'red')
}

plot(density(beta_0), main = 'Distribution of beta_0')
plot(density(beta_1), main = 'Distribution of beta_1')
```

The standard deviations of $\hat{\beta}_0$ and $\hat{\beta}_1$ are:

```r
sd(beta_0)
sd(beta_1)
```

The standard errors of $\beta_0$ and $\beta_1$ are as follows:

$SE(\hat{\beta}_0)^2 = \sigma^2\left[1/n+\frac{\bar{x}^2}{\sum_{i=1}^n\left(x_i-\bar{x}\right)^2}\right]$

$SE(\hat{\beta}_1)^2 = \frac{\sigma^2}{\sum_{i=1}^n\left(x_i-\bar{x}\right)^2}$

where $\sigma^2 = Var(\epsilon)$ that can be estimated as residual standard error (RSE):

$\sigma = RSE = \sqrt{RSS/(n-2)}$

95% confidence intervals are:
$\left[\hat{\beta}_0-2*SE(\hat{\beta}_0),\hat{\beta}_0+2*SE(\hat{\beta}_0)\right]$

$\left[\hat{\beta}_1-2*SE(\hat{\beta}_1),\hat{\beta}_1+2*SE(\hat{\beta}_1)\right]$

```r
lr <- lm(Sales ~ TV, data = adv)
beta_0_hat <- coef(lr)[1]
beta_1_hat <- coef(lr)[2]

y_hat <- beta_0_hat + beta_1_hat*adv$TV
RSS <- sum((adv$Sales - y_hat)**2)
n <- nrow(adv)
RSE <- sqrt(RSS/(n-2))
SE_beta_0 <- sqrt((RSE**2)*(1/n+mean(adv$TV)**2/(sum((adv$TV-mean(adv$TV))**2))))
SE_beta_1 <- sqrt((RSE**2)/(sum((adv$TV-mean(adv$TV))**2)))
RSE
SE_beta_0
SE_beta_1
```

... So, what are the possible values of $\beta_0$ and $\beta_1$? 

The 95\% confidence interval looks like this:

```r
beta_0_hat - 2*SE_beta_0
beta_0_hat + 2*SE_beta_0

beta_1_hat - 2*SE_beta_1
beta_1_hat + 2*SE_beta_1
```

In summary, in the absence of TV advertising, on average, the sales will be between 6,117 and 7,948 units. Every 1,000 increase in TV advertising will result in increase in sales of between 42 and 53 units.

### Hypothesis tests on the coefficients

There is relationship between TV and Sales, but is it significant?

H0: There is no relationship between TV and Sales
Ha: There is some relationship between TV and Sales

H0: $\beta_1 = 0$

Ha: $\beta_1 <> 0$

t-statistics:

$t = \frac{\hat{\beta}_1}{SE(\hat{\beta}_1)}$

Reject H0 if $\alpha$ < 0.05

```r
t_stat <- beta_1_hat/SE_beta_1
t_stat
```

```r
alpha <- 1-pnorm(t_stat)
alpha < 0.05
```

The null hypothesis ($\beta_1 = 0$) can be rejected, so $\beta_1 <> 0$, ergo there is a relationship between TV and Sales.
